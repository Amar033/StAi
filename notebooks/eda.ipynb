### EDA

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = [
    "AAPL", "GOOGL", "MSFT", "TSLA",
    "RELIANCE.NS", "TCS.NS", "INFY.NS", "WIPRO.NS",
    "^NSEI", "^BSESN", "^GSPC"
]
```

```python
fig, axes =plt.subplots (3,4,figsize=(15,15))
ax=axes.ravel()
for i in range(min(11,len(tickers))):
    ticker=tickers[i]
    df=pd.read_csv (f"../data/processed/{ticker}_processed.csv")
    ax[i].plot(df["Close"],label='Close Price')
    ax[i].set_title(f"{ticker} stock price over time")
    ax[i].set_xlabel('Days')
    ax[i].set_ylabel('Price ($)')
    ax[i].legend()
    ax[i].grid(True)

for j in range(len(tickers), 12):
    ax[j].set_visible(False)

plt.tight_layout()
plt.show()
plt.savefig('../assets/plots/stock_price.png')
```

```python
fig, axes =plt.subplots (4,3,figsize=(15,15))
ax=axes.ravel()
for i in range(min(11,len(tickers))):
    ticker=tickers[i]
    df=pd.read_csv (f"../data/processed/{ticker}_processed.csv")
    ax[i].hist(df["Returns"],bins=100)
    ax[i].set_title(f'Distribution of Daily Returns for {ticker}')
    
for j in range(len(tickers), 12):
    ax[j].set_visible(False)

plt.tight_layout()
plt.show()
plt.savefig('../assets/plots/daily_returns.png')
```

```python
fig, axes =plt.subplots (4,3,figsize=(15,15))
ax=axes.ravel()


for i in range(min(11,len(tickers))):
    ticker=tickers[i]
    df=pd.read_csv (f"../data/processed/{ticker}_processed.csv")
    sns.heatmap(df.corr(),annot=True,cmap='coolwarm',ax=ax[i])
    ax[i].set_title(f'Feature Correlation Matrix of{ticker}')
    
for j in range(len(tickers), 12):
    ax[j].set_visible(False)

plt.tight_layout()
plt.show()
plt.savefig('../assets/plots/Feature_correlation.png')
```
